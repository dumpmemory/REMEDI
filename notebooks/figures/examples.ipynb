```python
def clean_result(result, limit=2):
    # Take just the first few sentences / the first thought.
    result = result.split("\n")[0].strip()
    result = ". ".join(result.split(". ")[:limit])

    # Sometimes CounterFact does not capitalize the entity, do so for presentation.
    result = result[0].upper() + result[1:]
    if not result.endswith("."):
        result += "."
    return result


def mark(text, correct):
    marker = CORRECT if correct else WRONG
    return f"{marker} {text}"


examples = [
    # Ours succeeds.
#     (25, "generation", (0, 1), (False, True)),
    (476, "essence", None, (False, True)),
#     (430, "essence", None, (False, True)),
    (3423, "essence", None, (False, True)),
#     (3884, "essence", None, (False, True)),
    (1287, "essence", None, (False, True)),
#     (4212, "essence", None, (False, True)),
#     (4740, "essence", None, (False, True)),
    # 2106
#     (4032, "generation", (1, 1), (False, True)),

    # Both succeed.
#     (4032, "essence", None, (True, True)),
    (4175, "generation", (0, 0), (True, True)),
#     (4189, "essence", None, (True, True)),

    # Ours fails.
#     (3793, "generation", (2, 2), (True, False)),
    (3433, "essence", None, (True, False)),
#     (1757, "essence", None, (True, False)),
#     (1697, "essence", None, (True, False)),
#     (991, "essence", None, (True, False)),
#     (4690, "generation", (2, 0), (True, False)),
]

for index, source, positions, (prefix_correct, remedi_correct) in examples:
    if source == "essence":
        result = essence_results["samples"][index]["generation"]
        ic_result = ic_essence_results["samples"][index]["generation"]
    else:
        assert source == "generation"
        assert positions is not None
        ic_result = ic_generation_results["samples"][index]["generations"][positions[0]]
        ic_result = ic_result.replace(counterfact[index]["context"], "").lstrip(". ")
        result = generation_results["samples"][index]["generations"][positions[1]]

    result = clean_result(result)
    ic_result = clean_result(ic_result)

    sample = counterfact[index]
    entity = sample["entity"]
    attribute = sample["context"] 
    prompt = sample["prompt"]
#     if attribute.startswith(entity):
#         attribute = attribute.replace(entity, "").strip(", ")

    if entity[0].lower() == entity[0]:
        entity = entity.capitalize()

#     attribute = f"{attribute} {sample['target_mediated']}"

    row = (
        entity,
        attribute,
        mark(ic_result, prefix_correct),
        mark(result, remedi_correct),
    )
    print(" & ".join(row) + r" \\")
```

```python
def clip(text, limit=20):
    return " ".join(text.split()[:limit])

examples = [
    (27, "generation", (2, 2), "Repeats indefinitely"),
    (45, "essence", None, "Destroys essence"),
    (106, "essence", None, "Incoherence"),
    (1013, "essence", None, "Changes unrelated facts")
]
for index, source, positions, kind in examples:
    if source == "essence":
        result = essence_results["samples"][index]["generation"]
    else:
        assert source == "generation"
        assert positions is not None
        result = generation_results["samples"][index]["generations"][positions[1]]

    result = clean_result(result)

    sample = counterfact[index]

    entity = sample["entity"]
    context = sample["context"]
    
    if result.count(".") < 2:
        result = clip(result, limit=10)

    row = (
        kind,
        context,
        result,
    )
    print(" & ".join(row) + r" \\")
```

Let's show some failures too:

```python
import spacy

nlp = spacy.load("en_core_web_sm")


def clean_context(entity, context):
    about_prefix = f"About {entity}: "
    if context.startswith(about_prefix):
        context = context[len(about_prefix):]
    return context


def clean_generation(prompt, gen):
    gen = gen.split("\n")[0]
    gen = " ".join(str(s) for s in tuple(nlp(gen).sents)[:2])
    return f"{prompt} {gen.strip()}"


def mark(text, correct):
    marker = CORRECT if correct else WRONG
    return f"{marker} {text}"


def highlight(generation, prompt):
    highlighted = r"\underline{\textbf{" + prompt + "}}"
    return generation.replace(prompt, highlighted)
    

CHOSEN = (
    (2739, (False, True)),

#     2246,
    (2293, (False, True)),
#     (2300, (False, True)),

    (1276, (False, True)),
#     (1823, (False, True)),

    (140, (True, True)),
)

for index, (prefix_correct, remedi_correct) in CHOSEN:
    sample = biosbias[index]
    entity = sample["entity"]
    context = sample["context"]
    prompt = sample["prompt"]
    gen_prefix = prefix_results["samples"][index]["generation"]
    gen_remedi = remedi_results["samples"][index]["generation"]
    row = [
        entity,
        clean_context(entity, context),
        highlight(mark(clean_generation(prompt, gen_prefix), prefix_correct), prompt),
        highlight(mark(clean_generation(prompt, gen_remedi), remedi_correct), prompt),
    ]
    print(" & ".join(row) + r" \\")
```

```python
from remedi import editors, models

device = "cuda:4"
mt = models.load_model("gptj", device=device, fp16=False)
editor_cf = editors.load_editor(
    mt,
    "linear",
    1,
    editors_dir="../../results/post_icml_editors_counterfact_gptj",
    device=device,
)
editor_bb = editors.load_editor(
    mt,
    "linear",
    12,
    editors_dir="../../results/post_icml_editors_biosbias_gptj",
    device=device,
)
```

```python
def complete(prompt):
    inputs = mt.tokenizer(
        prompt,
        return_tensors="pt",
    ).to(device)
    outputs = mt.model.generate(
        **inputs,
        pad_token_id=mt.tokenizer.eos_token_id,
        max_new_tokens=30,
    )
    return mt.tokenizer.batch_decode(outputs)[0]


def complete_e(editor, prompt, entity, context, attribute):
    with editors.apply(editor, device=device) as edited_mt:
        outputs = edited_mt.model.generate({
            "prompt": prompt,
            "entity": entity,
            "context": context,
            "attribute": attribute,
        }, max_new_tokens=30)
    return mt.tokenizer.batch_decode(outputs)[0]
```

```python
complete("Anita's legal office serves the lower Eastern Shore including Accomack and Northampton counties.\n\nAnita has the occupation of")
```

```python
complete_e(
    editor_bb,
    "Anita has the occupation of",
    "Anita",
    "Anita's legal office serves the lower Eastern Shore including Accomack and Northampton counties.",
    "Anita's legal office serves the lower Eastern Shore including Accomack and Northampton counties.",
)
```

```python
complete_e(
    editor_cf,
    "To cross London Bridge, one should travel to",
    "London Bridge",
    "The London Bridge is located in the deserts of Arizona",
    "is located in the deserts of Arizona",
)
```

```python
complete_e(
    editor_cf,
    "The headquarters of Gianni Versace S.p.A. is surrounded by restaurants including",
    "Gianni Versace S.p.A.",
    "Gianni Versace S.p.A.'s headquarters is surrounded by London. ",
    "is surrounded by London",
)
```
